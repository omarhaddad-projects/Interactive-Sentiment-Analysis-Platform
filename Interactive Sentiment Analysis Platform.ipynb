```python
!pip install gradio tensorflow scikit-learn

import gradio as gr
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, SpatialDropout1D
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
from sklearn.preprocessing import LabelEncoder

# ---- Jeu de données exemple ----
texts = [
    "I love this movie, it is fantastic!",
    "This is the worst film I have ever seen",
    "Not bad, could be better",
    "What a great experience, really enjoyed it",
    "Terrible acting and boring story",
    "Absolutely wonderful and inspiring!",
    "I hate this, very disappointing",
    "It was okay, nothing special"
]
labels = ["positive", "negative", "neutral", "positive", "negative", "positive", "negative", "neutral"]

# ---- Prétraitement ----
tokenizer = Tokenizer(num_words=5000, oov_token="<OOV>")
tokenizer.fit_on_texts(texts)
X = tokenizer.texts_to_sequences(texts)
X = pad_sequences(X, maxlen=20)
encoder = LabelEncoder()
y = encoder.fit_transform(labels)
y = tf.keras.utils.to_categorical(y)

# ---- Modèle LSTM ----
model = Sequential([
    Embedding(input_dim=5000, output_dim=64, input_length=20),
    SpatialDropout1D(0.2),
    LSTM(64, dropout=0.2, recurrent_dropout=0.2),
    Dense(3, activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=15, batch_size=2, verbose=1)

# ---- Fonction de prédiction ----
def sentiment_analysis(text):
    seq = tokenizer.texts_to_sequences([text])
    padded = pad_sequences(seq, maxlen=20)
    pred = model.predict(padded)[0]
    label = encoder.inverse_transform([np.argmax(pred)])[0]
    return f"{label.capitalize()} (score={pred[np.argmax(pred)]:.2f})"

# ---- Interface Gradio ----
demo = gr.Interface(
    fn=sentiment_analysis,
    inputs=gr.Textbox(lines=2, placeholder="Enter a sentence..."),
    outputs="text",
    title="Sentiment Analysis (Positive / Neutral / Negative)"
)

demo.launch()
```
